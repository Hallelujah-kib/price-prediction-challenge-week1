## Correlation Between News Sentiment and Stock Movement

In this notebook, we analyze the relationship between news headline sentiment and daily stock returns.
Using TextBlob for sentiment analysis, we quantify the sentiment of news headlines,
aggregate daily sentiment scores, and correlate them with daily stock returns.

```python
import sys
import os

# Add project root to Python path
sys.path.append(os.path.abspath(os.path.join('..')))
```

```python
import pandas as pd
from src.sentiment_correlation import (
    load_news_data,
    preprocess_news_data,
    compute_sentiment_scores,
    aggregate_daily_sentiment,
    
    load_stock_returns,
    align_data,
    compute_correlation,
    plot_correlation,
)
```

## Set tickers and file paths

```python
tickers = ["AAPL", "AMZN"]

for ticker in tickers:
    print(f"Processing {ticker}...")

    news_path = f"../data/raw/raw_analyst_ratings.csv"
    stock_path = f"../data/processed/{ticker}_processed.csv"

    # Load and prepare news data
    news_df = load_news_data(news_path)
    news_df = preprocess_news_data(news_df)
    news_df = compute_sentiment_scores(news_df)
    daily_sentiment_df = aggregate_daily_sentiment(news_df)
    
    # Load stock returns
    stock_returns_df = load_stock_returns(stock_path)
    print(f"Stock returns colums: {stock_returns_df.columns} \n Daily Sentiment Columns: {daily_sentiment_df.columns}")
    # Align data by date
    merged_df = align_data(stock_returns_df, daily_sentiment_df)

    # Compute and print correlation
    corr = compute_correlation(merged_df)
    print(f"{ticker} correlation between sentiment and returns: {corr:.4f}")

    # Plot correlation
    plot_correlation(merged_df, ticker)
```
