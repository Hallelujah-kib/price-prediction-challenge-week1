## Quantitative Stock Analysis using Technical Indicators

In this notebook, we compute key technical indicators using TA-Lib and fetch financial metrics using PyNance for quantitative analysis of stock price data. The indicators provide insight into momentum, trends, and potential reversal points.

We modularize the code using a helper class to allow easy reuse for all 7 companies

```python
import sys
import os

# Add project root to Python path
sys.path.append(os.path.abspath(os.path.join('..')))

import pandas as pd
from src.utils import load_stock_data, save_processed_data
from src.indicators import StockIndicatorAnalyzer
```

```python
tickers = ["GOOG", "AMZN", "GOOG", 
           "MSFT", "AAPL", "TSLA", "NVDA"]

for ticker in tickers:
    input_path = f"../data/raw/{ticker}_historical_data.csv"
    df = load_stock_data(ticker, input_path)

    if not df.empty:
        analyzer = StockIndicatorAnalyzer(df)
        result_df = analyzer.compute_indicators()
        result_df = analyzer.add_daily_return(result_df)
        save_processed_data(result_df, ticker)
```
